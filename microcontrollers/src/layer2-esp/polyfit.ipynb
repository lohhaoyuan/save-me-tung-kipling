```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = """0 2.79
5 2.22
10 1.96
15 1.80
20 1.68
25 1.48
30 1.36
35 1.19
40 1.10
45 1.04
50 1.01
60 0.87
70 0.83
80 0.78
90 0.62
100 0.53
110 0.52
120 0.53
130 0.52
140 0.52
150 0.35
160 0.24
170 0.37
180 0.09
190 0.17
200 0.16"""
y = [float(p.split(" ")[0]) for p in s.split("\n")]
x = [float(p.split(" ")[1]) for p in s.split("\n")]

print("Input:")
for i in range(len(x)):
    print(x[i], y[i])
print()

print("Results:")
print("degree | mean error | polynomial coefficients")
for i in range(10):
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    y_pred = p(x)

    error = 0
    for j in range(len(x)):
        error += abs(y_pred[j] - y[j])
    me = error / len(x)
    print(i, me, poly, sep=" | ")


# Iterate through the first 5 polynomial degrees
for i in range(1, 6):
    # Fit the polynomial
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    x_sample = np.linspace(x[0], 0, 1000)
    y_pred = p(x_sample)

    # Plot the data and the fit
    plt.figure()
    plt.scatter(y, x)
    plt.plot(y_pred, x_sample, color="red")
    plt.title(f"Polynomial fit degree {i}")
    plt.xlabel("Y")
    plt.ylabel("X")
    plt.show()
```
