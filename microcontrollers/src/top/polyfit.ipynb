```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ball regress
s = """0 16
10 23
20 40
30 60
40 79
50 90
60 97
70 103
80 107
90 113
100 114
110 115
120 116
130 122
140 123
150 123
160 123
170 123
180 123
190 123
200 123"""
y = [float(p.split(" ")[0]) for p in s.split("\n")]
x = [float(p.split(" ")[1]) for p in s.split("\n")]

print("Input:")
for i in range(len(x)):
    print(x[i], y[i])
print()

print("Results:")
print("degree | mean error | polynomial coefficients")
for i in range(10):
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    y_pred = p(x)

    error = 0
    for j in range(len(x)):
        error += abs(y_pred[j] - y[j])
    me = error / len(x)
    print(i, me, poly, sep=" | ")


# Iterate through the first 5 polynomial degrees
for i in range(1, 11):
    # Fit the polynomial
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    x_sample = np.linspace(0, 140, 1000)
    y_pred = p(x_sample)

    # Plot the data and the fit
    plt.figure()
    plt.scatter(y, x)
    plt.plot(y_pred, x_sample, color="red")
    plt.title(f"Polynomial fit degree {i}")
    plt.xlabel("Y")
    plt.ylabel("X")
    plt.show()
```
