```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ball regress
s = """7 64
12 73
17 82
22 91
27 98
32 104
37 108
42 113
47 117
60 125
70 129
80 133
90 137
100 139
110 141
120 143
130 145
140 144
150 146"""
y = [float(p.split(" ")[0]) for p in s.split("\n")]
x = [float(p.split(" ")[1]) for p in s.split("\n")]

print("Input:")
for i in range(len(x)):
    print(x[i], y[i])
print()

print("Results:")
print("degree | mean error | polynomial coefficients")
for i in range(10):
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    y_pred = p(x)

    error = 0
    for j in range(len(x)):
        error += abs(y_pred[j] - y[j])
    me = error / len(x)
    print(i, me, poly, sep=" | ")


# Iterate through the first 5 polynomial degrees
for i in range(1, 11):
    # Fit the polynomial
    poly = np.polyfit(x, y, i)
    p = np.poly1d(poly)
    x_sample = np.linspace(0, 140, 1000)
    y_pred = p(x_sample)

    # Plot the data and the fit
    plt.figure()
    plt.scatter(y, x)
    plt.plot(y_pred, x_sample, color="red")
    plt.title(f"Polynomial fit degree {i}")
    plt.xlabel("Y")
    plt.ylabel("X")
    plt.show()
```
